### Filtering data

```python
# Filtering the data based on more than one condition.

# AND condition
condition_df_1 = pokemon_data.loc[(pokemon_data['Type 1'] == 'Grass') & (pokemon_data['Type 2'] == 'Poison')]
```

```python
# OR condition

condition_df_2 = pokemon_data.loc[(pokemon_data['Type 1'] == 'Grass') | (pokemon_data['Type 2'] == 'Poison')]
```

```python
condition_df_3 = pokemon_data.loc[(pokemon_data['Type 1'] == 'Grass') & (pokemon_data['Type 2'] == 'Poison') & (pokemon_data['HP'] > 70)]
```

```python
condition_df_3
```

```python
# Reset the index for a dataframe.

condition_df_3.reset_index(drop=True, inplace=True)
```

```python
condition_df_3
```

```python
# Filter out the names that contain 'Mega.' Returns a dataframe with
# names that do not contain 'Mega.'

pokemon_data.loc[~pokemon_data['Name'].str.contains('Mega')]
```

```python
# Return a dataframe of names that contain 'Mega.'

pokemon_data.loc[pokemon_data['Name'].str.contains('Mega')]
```

### Implementing Regular Expressions

```python
import re
```

```python
# Simple way to determine if Type 1 is either 'Grass' or 'Fire'.
# Use a regular expression.

# pokemon_data.loc[pokemon_data['Type 1'].str.contains('Grass|Fire', regex=True)]

# To ignore case...

pokemon_data.loc[pokemon_data['Type 1'].str.contains('grass|fire', flags=re.I, regex=True)]
```

```python
# Just want data in the 'Name' column that starts with 'Pi'.
                                                # Start with:^
pokemon_data.loc[pokemon_data['Name'].str.contains('^pi[a-z]*', flags=re.I, regex=True)]
```

### Conditional changes

```python
# Change Type 1 = 'Fire' to 'Flamer'.

pokemon_data.loc[pokemon_data['Type 1'] == 'Fire', 'Type 1'] = 'Flamer'
```

```python
# We see that 'Fire' was changed to 'Flamer.'

pokemon_data.head(10)
```

```python
# To revert back to 'Fire'...

pokemon_data.loc[pokemon_data['Type 1'] == 'Flamer', 'Type 1'] = 'Fire'
```

```python
pokemon_data.head(10)
```

```python
# If the 'Total' column is greater than 500, the 'Generation' and 'Legendary'
# columns should be set to the value 'TEST VALUE.'

pokemon_data.loc[pokemon_data['Total'] > 500, ['Generation', 'Legendary']] = 'TEST VALUE'
```

```python
# To modify individually...
# The value in the 'Generation' column becomes 'Test 1' and the
# value in the 'Legendary' column becomes 'Test 2' should the first
# condition be true.

pokemon_data.loc[pokemon_data['Total'] > 500, ['Generation', 'Legendary']] = ['Test 1', 'Test 2']
```

### Using (groupby) for aggregate statistics

```python
# Using mean...

pokemon_data.groupby(['Type 1']).mean().sort_values('Defense', ascending=False)
```

```python
# Using count...

# Add a 'Count' column to the dataframe.

pokemon_data['Count'] = 1
```

```python
pokemon_data.head(5)
```

```python
# Adding the Count column makes it easier to see the number
# of Pokemon for each Type 1.

pokemon_data.groupby(['Type 1']).count()['Count']
```

```python
# Look at Type 1 and Type 2.

pokemon_data.groupby(['Type 1', 'Type 2']).count()['Count']
```

```python
# Loading in a dataframe in chunks...

# The chunksize specified is the number of rows to be
# included in each dataframe.

count = 0
for df in pd.read_csv('pokemon_data.csv', chunksize=5):
    count += 1
    print('This is df #:', count)
    print(df)
```
